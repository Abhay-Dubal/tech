```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('zomato.csv')
```

```python
df
```

```python
print(" get detailed Info ")
df.info() 
```

```python
print("differtiating data types")
obj = []
others = []

for col in df.columns:
    if df[col].dtypes == 'object':
        obj.append(col)
    else:
        others.append(col)
        
print("object data types")
print(obj)
print("***************************************************************************************")
print("int , float , boolean data types")
print(others)
```

```python
df.describe()
```

```python
print("Checking null values")

print("getting only those column having atleast one null value")
for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(f'column_name = {col} , Total_null_values = ',df[col].isnull().sum())
```

```python
s = sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')


s.set_xlabel('Column Names', fontsize=30)
s.set_ylabel('Null Count', fontsize=30)
```

```python
df_new = pd.read_excel('Country-Code.xlsx')
df_new
```

```python
df_merge = pd.merge(df,df_new,on='Country Code')
df_merge.sample(4)
```

```python
print("Getting Unique values Count")



for col in df_merge.columns:

    print(f'column_name = "{col}" , Total_unique_values = [{df_merge[col].value_counts().count()}]')
```

```python
df = df_merge
```

```python
#  Plotting pie chart for column values
col_name = 'Country'

#  getting unique values names
unique_values = df[col_name].value_counts().index

#  getting respective counts
unique_values_count = df[col_name].value_counts().values


plt.pie(unique_values_count,labels=unique_values,autopct='%1.1f%%',shadow=True,startangle=90 )
plt.title("your title")
plt.legend(title = "Title here")
plt.show() 
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(unique_values_count,labels=unique_values,autopct='%1.1f%%',shadow=True,startangle=90 )
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.legend(title = "Title here")
plt.title("your title")
plt.show()
```

```python
# considering first n values

col_name = 'Country'

consider_first_n_values = 5   # start from 1 
#  getting unique values names
unique_values = df[col_name].value_counts().index

unique_values = unique_values[:consider_first_n_values]
#  getting respective counts
unique_values_count = df[col_name].value_counts().values
unique_values_count = unique_values_count[:consider_first_n_values]

plt.pie(unique_values_count,labels=unique_values,autopct='%1.1f%%',shadow=True,startangle=90 )
plt.legend(title = "Title here")
plt.title("your title")
plt.show() 
```
